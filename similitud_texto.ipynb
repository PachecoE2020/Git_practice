```python
import gensim.downloader as api
model = api.load("glove-wiki-gigaword-50")
print("Tamaño del vocabulario:", len(model))
print("Dimensión de los embeddings:", model.vector_size)
```

```python
import numpy as np
from numpy.linalg import norm

sentences = [
    "el gato duerme en la cama",
    "el perro duerme en la alfombra",
    "el gato come pescado",
    "el perro come carne",
    "el gato juega con la pelota",
    "el perro juega en el parque",
    "el coche corre por la carretera",
    "el camión circula por la autopista",
    "el coche tiene ruedas",
    "el camión transporta carga",
    "el médico trabaja en el hospital",
    "la enfermera trabaja en el hospital",
    "el doctor atiende pacientes",
    "la enfermera cuida pacientes"
]

def get_sentence_embedding(sentence):
    words = sentence.lower().split()
    embeddings = []
    for word in words:
        if word in model:
            embeddings.append(model[word])
    if embeddings:
        return np.mean(embeddings, axis=0)
    else:
        # Return a zero vector if no words are found in the model vocabulary
        return np.zeros(model.vector_size)

def cosine_similarity(vec1, vec2):
    if norm(vec1) == 0 or norm(vec2) == 0:
        return 0.0 # Handle cases where one or both vectors are zero vectors
    return np.dot(vec1, vec2) / (norm(vec1) * norm(vec2))

print("Functions 'get_sentence_embedding' and 'cosine_similarity' created.")
```

```python
embeddings = []
for sentence in sentences:
    sentence_embedding = get_sentence_embedding(sentence)
    print(f"Sentence: {sentence}")
    print(f"Embedding: {sentence_embedding}")
    embeddings.append(sentence_embedding)
```

```python
import pandas as pd

similarity_matrix = np.zeros((len(sentences), len(sentences)))

for i in range(len(sentences)):
    for j in range(len(sentences)):
        if i <= j: # Calculate for upper triangle to avoid redundant calculations and ensure symmetry
            sim = cosine_similarity(embeddings[i], embeddings[j])
            similarity_matrix[i, j] = sim
            similarity_matrix[j, i] = sim # The similarity matrix is symmetric

similarity_df = pd.DataFrame(similarity_matrix, index=sentences, columns=sentences)
print("Matriz de Similitud de Coseno entre Oraciones:")
display(similarity_df.round(4))
```
